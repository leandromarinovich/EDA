#2. An√°lisis gr√°ficos üìä

**Para comenzar el an√°lisis, se selecciona la columna `sector` con el fin de examinar c√≥mo se encuentran distribuidas las obras.**

```python
print('Valores √∫nicos de la columna "sector"\n')
df['sector'].unique()
```

```python
# Crear un gr√°fico de barras para analizar la diferencia entre las cantidades de obras de los diversos sectores
df['sector'].value_counts().plot(kind='barh')


plt.xlabel('Sector')
plt.ylabel('Cantidad')
plt.title('Cantidad de obras seg√∫n Sector')


plt.show()
```

```python
# Calculamos la suma del gasto por sector para generar un gr√°fico de barras
media_gasto = df.groupby('sector')['monto_total'].sum()

# Crear el gr√°fico de barras (dividimos por un bill√≥n para que se entienda mejor)
media_gasto.div(1000000000).plot(kind='barh')

plt.xlabel('Sector')
plt.ylabel('Gasto (en billones de pesos)')
plt.title('Gasto por sector')

plt.show()
```

```python
# Calculamos la media del gasto por sector para generar un gr√°fico de barras
media_gasto = df.groupby('sector')['monto_total'].mean()

# Crear el gr√°fico de barras (dividimos por un mill√≥n para que se entienda mejor)
media_gasto.div(1000000).plot(kind='barh')

plt.xlabel('Sector')
plt.ylabel('Media del gasto (en millones de pesos)')
plt.title('Media del gasto por sector')

plt.show()
```

**Del an√°lisis por `sector` se observa que las obras de Cloacas son, por amplio margen, las m√°s numerosas. Sin embargo, no representan una gran erogaci√≥n de dinero, ya su media de gasto es de las m√°s bajas. En contraste, las vinculadas a Recursos H√≠dricos demandan una inversi√≥n considerable. Llama la atenci√≥n la escasa asignaci√≥n destinada a Vivienda, tanto en n√∫mero de proyectos como en montos. Educaci√≥n y Salud, otros dos pilares, tambi√©n son muy bajos**

**Ahora, buscamos profundizar seg√∫n el Tipo de Proyecto. Para ello, graficamos un Top 10 de tipos m√°s utilizados**

```python
df_por_tipo_proyecto = df.groupby('tipo_proyecto').count()['id_proyecto'].sort_values(ascending=False).head(10)


# Gr√°fico de barras horizontales
ax = df_por_tipo_proyecto.plot(kind='barh', figsize=(10,6))

# Invertir el eje Y para que el mayor quede arriba
ax.invert_yaxis()

# Etiquetas en cada barra
for i, v in enumerate(df_por_tipo_proyecto):
    ax.text(v + 0.5, i, str(v), va='center')

# Est√©tica del gr√°fico
plt.xlabel('Cantidad')
plt.ylabel('Tipo de Proyecto')
plt.title('Top 10 de Tipos de Proyectos realizados')
plt.tight_layout()
plt.show()
```

```python
#Hacemos un group by para ver a qu√© sector corresponde cada tipo de proyecto

df_tipo_sector = (
    df.groupby(['tipo_proyecto', 'sector'])['id_proyecto']
      .count()
      .reset_index()
      .sort_values('id_proyecto', ascending=False)
).rename(columns={'id_proyecto': 'total'}).head(10)



df_tipo_sector.head(10)
```

**Por otro lado, analizamos qu√© provincias se vieron m√°s beneficiadas por las obras**

```python
#Busco los valores √∫nicos de provincias
df['provincia'].unique()
```

```python
#Genero un gr√°fico de torta para entender c√≥mo se distribuyen. La ide era un mapa interactivo pero no encontr√© ning√∫n GEOJson para hacer coordenadas

# Contar obras por provincia
df_provincias = df['provincia'].value_counts()

# Top 10 + resto como "Otras" porque -si no- el gr√°fico no se entender√≠a
top10 = df_provincias.head(10)
otras = pd.Series({'Otras': df_provincias.iloc[10:].sum()})
df_grafico = pd.concat([top10, otras])


palette = sns.color_palette("Set3", n_colors=len(df_grafico))

plt.figure(figsize=(8,8))
plt.pie(df_grafico, labels=df_grafico.index, autopct='%1.1f%%', startangle=90, colors=palette)
plt.title('Top 10 provincias por cantidad de obras')
plt.show()

```

**Para continuar con el an√°lisis, vemos un gr√°fico de l√≠neas temporales en donde podemos observar cu√°ntas obras se fueron realizando seg√∫n a√±o. Primero vemos los valores √∫nicos para entender desde cu√°ndo se contabiliza.**

```python
df['anio_inicio'].unique()
```

```python
df_por_anio = df.groupby('anio_inicio').count()['id_proyecto'].sort_index()

plt.figure(figsize=(10,6))
plt.plot(df_por_anio.index, df_por_anio.values, marker='o', linestyle='-', color='b')
plt.title('Cantidad de proyectos iniciados por a√±o')
plt.xlabel('A√±o de inicio')
plt.ylabel('Cantidad de proyectos')
plt.grid(True)
plt.tight_layout()
plt.show()
```

**Como vemos que seguramente hay datos faltantes entre 2008 y 2015, posiblemente es mejor pensar el gr√°fico a partir del a√±o finalizado o a finalizar**

```python
df_por_anio = df.groupby('anio_fin').count()['id_proyecto'].sort_index()

plt.figure(figsize=(10,6))
plt.plot(df_por_anio.index, df_por_anio.values, marker='o', linestyle='-', color='b')
plt.title('Cantidad de proyectos finalizados o a finalizar por a√±o')
plt.xlabel('A√±o de finalizaci√≥n')
plt.ylabel('Cantidad de proyectos')
plt.grid(True)

plt.tight_layout()
plt.show()
```

**Tambi√©n buscamos analizar c√≥mo van esas obras de avanzadas, tanto financiera como f√≠sicamente. Primero las analizamos por separado y luego hacemos una comparaci√≥n entre ambas.**

```python
df['avance_financiero'].hist(bins=5)

# Configurar las etiquetas del eje x y el t√≠tulo del gr√°fico
plt.xlabel('Avance Financiero')
plt.ylabel('Conteo')
plt.title('Avance Financiero de las Obras en Rangos de 10')

# Mostrar el gr√°fico
plt.show()
```

```python
df['avance_fisico'].hist(bins=5)

# Configurar las etiquetas del eje x y el t√≠tulo del gr√°fico
plt.xlabel('Avance F√≠sico')
plt.ylabel('Conteo')
plt.title('Avance F√≠sico de las Obras en Rangos de 10')

# Mostrar el gr√°fico
plt.show()
```

```python
# Transformamos el Dataframe a formato largo para luego generarle una nueva categor√≠a seg√∫n d√≥nde se encuentre el avance
df_unido = df.melt(value_vars=['avance_financiero','avance_fisico'],
                  var_name='tipo', value_name='avance')

df_unido['rango'] = pd.cut(df_unido['avance'], bins=5, labels=['0-20','20-40','40-60','60-80','80-100'], include_lowest=True)

# Gr√°fico de barras comparativo

plt.figure(figsize=(10,6))
sns.histplot(data=df_unido, x='rango', hue='tipo', multiple='dodge', shrink=0.9)
plt.ylabel('Cantidad de obras')
plt.xlabel('Rango de avance (%)')
plt.title('Comparaci√≥n de avance financiero vs f√≠sico')
plt.show()
```

**En el an√°lisis, podemos ver c√≥mo entre el 80 y 100% hay muchas obras ya terminadas o por terminarse que a√∫n no se han terminado de pagar. Por el contrario, en el resto de rangos el avance financiero es superior al f√≠sico. Se pag√≥ m√°s de lo que se est√° avanzando en la obra en s√≠.**

**Por otro lado, a continuaci√≥n mostramos c√≥mo queda el Dataframe unido. Las columnas de `avance_financiero` y `avance_fisico` se convirtieron en valores categ√≥ricos y se sumaron dos columnas de `avance` y en qu√© `rango` est√° este avance**
